```python
import csv

with open('step1.csv', newline='') as f:
    reader = csv.reader(f)
    data = list(reader)

length=len(data)
mylist=[]
for i in range(length):
    mylist.append(int(data[i][0])*3+6)
mylist
```

```python
import numpy as np
np.savetxt("step2.csv", mylist, delimiter=",", fmt='%s')
```
